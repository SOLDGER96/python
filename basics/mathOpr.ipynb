```python
int1 = 56
int2 = 44

print(f"Addition: {int1 + int2}")
print(f"Subtraction: {int1 - int2}")
print(f"Product: {int1 * int2}")
print(f"Divison: {int1 / int2}")
print(f"Remainder: {int1 % int2}")
print(f"Floor Division: {int1 // int2}")

int1 = 2
int2 = 3
print(f"Ans: {int1 ** int2}") # 2x2x2 OR 2^3
```

```python
# int + float
int1 = 45
float1 = 43.4

print(f"Addition: {int1 + float1}, Type: {type(int1+float1)}")
```

```python
# Conversion
float1 = 44.45
int1 = 44
conFlo = float(int1)
conInt = int(float1)
print(conInt)
print(conFlo)
```

```python
# Round up numbers
int1 = 88.758
int1 = round(int1)
print(int1)
```

```python
# Absolute value
num = -16
num = abs(num)
print(num)
```

```python
# Power
r = pow(2, 3) # 2x2x2 OR 2^3 OR 2 ** 3
print(r)
```
