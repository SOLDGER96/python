```python
# Integer
integer = 10
print("Interger value: ", integer)

# Float
floatvar = 9.6
print("Float var: ", floatvar)

# String 
stringvar = 'soldger96'
print('String value: ', stringvar)

# Boolean
myBool = True
print("Bool value: ", myBool)
```

```python
# Dictionary: A collection of key-value pairs enclosed in curly braces
myDict = {'name': 'Tanmay', 'IGN': 'soldger96'}
print('Dictionary: ', myDict)

# Set: An unordered collection of unique elements, like {4, 2, 0}
mySet = {7, 5, 4}
print('Set Values: ', mySet)

# Tuple: An immutable ordered collection, enclosed in brackets
myTuple = (3, 5, 7)
print("Tuple: ", myTuple)

# Range
myRange = range(7)
print('Range: ',myRange)

# List: An ordered collection of elements that supports different data types.
myList = [22, 'Tanmay', 5.8, True]
print('List: ', myList)
```

```python
# Get the type of any variable
print(type(integer))
print(type(floatvar))
print(type(stringvar))
print(type(myList))
print(type(myTuple))
```

```python
name = 'Alice'
print(name, type(name))

is_student = True
print(is_student, type(is_student))

age = 20
print(age, type(age))

score = 80.5
print(isinstance(score, float))
print(score, type(score))
```
