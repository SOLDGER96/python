## Example 1

#### Instance attributes are unique to each object created from a class, and you usually set them with the __init__ method. Class attributes, on the other hand, belong to the class itself and are shared by all instances of that class.

```python
class Dog:
    species = "French BullDog"  # Class attribute

    def __init__(self, name):
        self.name = name # Instance attribute

print(Dog.species)
```

```python
a = Dog("Bunny")
print(a.name)
print(a.species)
```

```python
b = Dog("Gunny")
print(a.name)
print(a.species)
```

## Example 2

```python
class Car:
    def __init__(self, color, model):
        self.color = color
        self.model = model

car1 = Car('Red', 'Lamborgini')
print(car1.model)
print(car1.color)
```

```python
car2 = Car('Blue', 'Urus')
print(car2.model)
print(car2.color)
```

## Example 3

```python
class Bike:
    def __init__(self, color, model):
        self.color = color
        self.model = model
    def describe(self):
        return f'This Bike is a {self.color} {self.model}'

b1 = Bike('Red', 'Pulsur')
print(b1.describe()) 
```
