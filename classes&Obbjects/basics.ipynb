#### It is common in Python to use the PascalCase convention when naming classes.

```python
class EmpInfo():
    def __init__(self, name, age):
        self.name = name
        self.age = age
    
    def sample_method(self):
        print(self.name.upper())
```

```python
class Dog:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def bark(self):
        print(f"{self.name.title()} says woof woof! I'm {self.age} years old")

a = Dog('Butter', 5)
b = Dog('Brownie', 2)
a.bark()
b.bark()
```

#### In summary, the difference between a class and an object is that a class is the template or the blueprint, and an object is what is created using that template.
