````python
class Car:
    def __init__(self, brand, car):
        self.brand = brand
        self.car = car
````

````python
my_car = Car('Lamborghini', 'Gallardo') 
print(my_car.brand) 
print(my_car.car) 
````

#### Example 1: getattr

````python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

person = Person('Holder', 34)

print(getattr(person, 'name'))
print(getattr(person, 'age'))
````

````python
print(getattr(person, 'city', 'ratnagiri'))
````

````python
print(getattr(person, 'city'))
````

#### the real power of getattr() is apparent when the attribute name comes from a variable, such as from user input or some file.

#### In that case, you can't use the regular object.attribute_name syntax because the attribute name is not fixed.

````python
person = Person('ayush', 22)

atrr_name = input('Enter attr to view: ')
print(getattr(person, atrr_name, 'Attribute not found!'))
````

#### In this case, if the user types in name, they see ayush, and if they type in age, they see 22. And if they type something that doesn't exist in the class like email, they see Attribute not found.

````python
person = Person('vasu', 23)

for attr in dir(person):
    if not attr.startswith('__') and not callable(getattr(person, attr)):
        value = getattr(person, attr)
        print(f'{attr}: {value}')
````

#### The setattr() function lets you create a new attribute or update an existing one dynamically. The syntax looks like this:
```bash
setattr(object, attribute_name, value)
```
