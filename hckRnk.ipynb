```python
def swap_case(s):
    s = list(s)
    arr = []
    for i in s:
        if i == i.upper():
            arr.append(i.lower())
        elif i == i.lower():
            arr.append(i.upper())
        string = "".join(arr)
            
    return string

if __name__ == '__main__':
    s = 'HackerRankJOd. o'
    result = swap_case(s)
    print(result)
```

```python
def mutate_string(string, position, character):
    swpChar = list(string)
    swpChar[position] = character
    s_new = ''.join(swpChar)
    return s_new

if __name__ == '__main__':
    s = input()
    i, c = input().split()
    s_new = mutate_string(s, int(i), c)
    print(s_new)
```

```python
def count_substring(string, sub_string):
    lostr = len(string)
    losstr = len(sub_string)
    count = 0 
    for i in range(lostr - losstr + 1):
        if string[i:i+losstr] == sub_string:
            count += 1
    return count

if __name__ == '__main__':
    string = input().strip()
    sub_string = input().strip()
    
    count = count_substring(string, sub_string)
    print(f'Count of {sub_string} in {string}: {count}')
```
