```python
# Only same type of variable can be concated
str1 = 'hello'
str2 = 'world'

coCatStr = str1 + ' ' + str2
print(coCatStr)
```

```python
name = 'soodger'
age = 20

namAge = name + str(age)
print(namAge)

# Appending into a string using '+='
namAdd = name
namAdd += str(age)
print(namAdd)
```
