```python
#A string is a sequence of characters surrounded by either 
# single or double quotation marks. 
# In some programming languages, characters 
# surrounded by single quotes are treated differently 
# than characters surrounded by double quotes, 
# but in Python, they're treated equally. 
# So, you can use either when working with strings. 
# Here are some examples of strings:
```

```python
str1 = 'Hello'
str2 = "Hello"

# Multiline String
str3 = '''Multiline
string'''
str4 = """this also
works """
```

```python
# Using ' and " in line
print("It's a sunny day.")
print('"Hello World!"')

# using \' 
print('It\'s a sunny day!')
```

```python
# Checking the presence of char in String
str5 = "Tanmay JOD"

print('1.','Tanmay' in str5)
print('2.', 'JOD' in str5)
print('3.', 'JOd' in str5)

# Printig length of string
print(len(str5))
```
