##### Nested Loops

```python
boy = ['agnelo', 'vasudev', 'ayush']
boy2 = ['atharva', 'pranav']

for b in boy:
    for b2 in boy2:
        print(b, b2)
```

##### nother type of loop you can use in Python is the while loop. This type of loop will repeat a block of code until the condition is False. 

```python
secNum = 4
guess = 0
while guess != secNum:
    guess = int(input("Guess a no b/w 1-5: "))
    if guess != secNum:
        print('Wrong! Try again.')
print('You got it.')
```

##### The break statement is used to stop the execution of a loop.

```python
names = ['agnelo', 'vasudev', 'pranav', 'ayush']

for n in names:
    if n == 'ayush':
        break
    print(n)
```

##### The continue statement is used to skip the current iteration of a loop and move onto the next iteration.

```python
names = ['agnelo', 'vasudev', 'pranav', 'ayush']

for n in names:
    if n == 'vasudev':
        continue
    print(n)
```

##### Both for and while loops can be combined with an else clause, which is executed only when the loop is not terminated by a break statement.

```python
words = ['sky', 'apple', 'rhythm', 'fly', 'orange']

for w in words:
    for letter in w:
        if letter.lower() in 'aeiou':
            print(f"'{w}' contain a vowel {letter}")
            break
    else:
        print(f"'{w}' has no vowel")
```
