```python
even = []

for num in range(21):
    if num % 2 == 0:
        even.append(num)

print(even)
```

```python
even = [num for num in range(21) if num % 2 == 0]
print(even)
```

```python
numbers = [1,2,3,4,5,6]
res = [(num, 'even') if num % 2 == 0 else (num, 'odd') for num in numbers]
print(res)
```

```python
words = ['tree', 'sky', 'mountain', 'river', 'cloud', 'sun']

def longWords(word):
    return len(word) > 4

longWord = list(filter(longWords, words))
print(longWord)
```

```python
celsius = [0, 10, 20, 30, 40]

def toFaren(temp):
    return (temp * 9/5) + 35

infaren = list(map(toFaren, celsius))
print(infaren)
```

```python
numbers = [5, 10, 15, 20]
total = sum(numbers)
print(total)
```

```python
numbers = [5, 10, 15, 20]
total = sum(numbers, 10) # positional argument
print(total)
```

```python
numbers = [5, 10, 15, 20]
total = sum(numbers, start=20) # keyword argument
print(total)
```
