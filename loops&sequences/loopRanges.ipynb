##### The range() function is used to generate a sequence of integers.

```python
for n in range(3):
    print(n)
```

##### range(start, stop, step)

```python
for n in range(1, 5):
    print(f'Number b/w 1 to 5: {n}')
```

```python
for n in range(2, 21, 2):
    print(f'table of 2: {n}')
```

```python
for n in range(40, 0, -10):
    print(n)
```

```python
numbers = list(range(2, 11, 2))
print(numbers)
```
