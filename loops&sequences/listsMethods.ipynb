 ##### In this lesson, you will continue to learn about lists and some common methods associated with them like append(), pop(), and sort().

##### Using append() method

```python
num = [1,2,3,4,5]
num.append(6)
num
```

##### Adding one list at the end of another

```python
num = [1,2,3,4,5]
numExtra = [6,7,8,9]
num.append(numExtra)
print(num)
```

##### But if you want to add all of the individual numbers from the even_numbers list at the end of the numbers list, then you can use the extend() method.

```python
num = [1,2,3,4,5]
numExtra = [6,7,8,9]

num.extend(numExtra)
print(num)
```

##### To insert an element at a specific index in a list, you can use the insert() method. This method accepts two arguments: the index where you wish to insert the new item and the item you want to insert.

```python
num = [1,2,3,4,5]
num.insert(2, 2.5)
num
```

##### to remove an element from a list, you can use the remove() method. The remove() method takes the value of the element to remove as an argument:

```python
num = [10, 20, 30, 40, 50]
num.remove(40)
num
```

##### To remove an element at a specific index in the list, you can use the pop() method like this:

```python
num = [10, 20, 30, 40, 50]
num.pop(2)
num
```

##### to empty the list, then you can use the clear() method like this:

```python
num = [10, 20, 30, 40, 50]
num.clear()
num
```

##### sort() method. This method is used to sort the elements in place.

```python
numbers = [19, 2, 35, 1, 67, 41]
numbers.sort()
numbers
```

```python
numbers = [19, 2, 35, 1, 67, 41]
sortNum = sorted(numbers)
print(sortNum)
```

##### reverse() method. This method, will reverse a list of elements in place like this:

```python
num = [6,5,4,3,2,1,0]
num.reverse()
num
```

##### index method. This is used to find the first index where an element can be found in a list.

```python
lang = ['python', 'rust', 'c++', 'go']
lang.index('c++')
```
