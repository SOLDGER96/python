##### using the count() method to count the number of times the element occured in a tuple

```python
num = (1,2,3,4,5,3,2,4,4,1)
print(num.count(4))

name = ('alice', 'agnelo', 'agnelo', 'ayush', 'agnelo')
print(f'agnelo repeated: {name.count('agnelo')} times')
```

##### index() method. This method is used to find the index where a particular item is present in a tuple. Here is an example of using the index() method to find the index for the string

```python
name = ('alice', 'agnelo', 'hari', 'agnelo', 
        'ayush', 'agnelo')
print(name.index('agnelo'))
print(name.index('agnelo', 2))
print(name.index('agnelo', 4, 6))
```

##### sorted() function can be used on any iterable including tuples.

```python
num = (2,5,3,6,7,3,4,8,9,2,5,6,3,4,0,4,3)
numSort = sorted(num)
print(numSort)
```

##### If you need to customize the sorting behavior for an iterable, you can use the optional reverse and key arguments. Here is an example of using key argument to sort items in a tuple by length:

```python
programming_languages = ('Rust', 'Java', 'Python', 'C++', 'Rust', 'Python')
sorted(programming_languages, key=len)
```

##### to create a new list of values in reverse order, then you can use the reverse argument like this:

```python
programming_languages = ('Rust', 'Java', 'Python', 'C++', 'Rust', 'Python')
print(sorted(programming_languages, reverse=True))
```
