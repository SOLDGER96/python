```python
if __name__ == '__main__':
    records = []
    for _ in range(int(input())):
        name = input()
        score = float(input())
        
        records.append([name, score])
    # print(records)
    scores = []
    for i, j in records:
        scores.append(j)
    # print(scores)
    scores = set(scores)
    print(scores)
    scores = list(scores)
    scores.sort()
    print(scores)
    secLow = scores[1]
    print(secLow)
    arr = []
    for i, j in records:
        if j == secLow:
            arr.append(i)
    
    arr.sort()
    for i in arr:
        print(i)
```
