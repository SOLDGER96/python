```python
def validate_isbn(isbn, length):
    # Fix 1: Corrected len() check (removed the extra argument)
    if len(isbn) != length:
        print(f'ISBN-{length} code should be {length} digits long.')
        return

    # Fix 2: Handle non-numeric characters with a try-except block
    try:
        # ISBN-10 allows 'X' at the end, ISBN-13 is all numbers
        if length == 10:
            # Check first 9 digits are numbers
            main_digits_list = [int(digit) for digit in isbn[:9]]
            given_check_digit = isbn[9].upper()
            expected_check_digit = calculate_check_digit_10(main_digits_list)
        else:
            # Check all 13 digits are numbers
            main_digits_list = [int(digit) for digit in isbn[:12]]
            given_check_digit = isbn[12]
            expected_check_digit = calculate_check_digit_13(main_digits_list)
            
    except ValueError:
        print('Invalid character was found.')
        return
    except IndexError:
        # This prevents crashing if the string is too short despite the len check
        print('Invalid ISBN Code.')
        return

    # Fix 3: Compare digits
    if given_check_digit == expected_check_digit:
        print('Valid ISBN Code.')
    else:
        print('Invalid ISBN Code.')

def calculate_check_digit_10(main_digits_list):
    digits_sum = 0
    for index, digit in enumerate(main_digits_list):
        digits_sum += digit * (10 - index)
    
    result = (11 - (digits_sum % 11)) % 11
    if result == 10:
        return 'X'
    return str(result)

def calculate_check_digit_13(main_digits_list):
    digits_sum = 0
    for index, digit in enumerate(main_digits_list):
        if index % 2 == 0:
            digits_sum += digit * 1
        else:
            digits_sum += digit * 3
    
    result = (10 - (digits_sum % 10)) % 10
    return str(result)

def main():
    user_input = input('Enter ISBN and length: ')
    
    # Fix 4: Handle missing comma (IndexError)
    try:
        values = user_input.split(',')
        if len(values) < 2:
            raise IndexError
        
        isbn = values[0].strip()
        
        # Fix 5: Handle non-numeric length (ValueError)
        try:
            length = int(values[1].strip())
        except ValueError:
            print('Length must be a number.')
            return

        # Fix 6: Indentation and logic flow
        if length == 10 or length == 13:
            validate_isbn(isbn, length)
        else:
            print('Length should be 10 or 13.')

    except IndexError:
        print('Enter comma-separated values.')

# Important: Commented out for the tests to run properly
# main()
```
