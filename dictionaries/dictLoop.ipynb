#### You can loop over a dictionary if you need to access and process its key-value pairs. This is helpful for updating their values or applying some logic to them.

```python
product = {
    'laptop': 63000,
    'smartphone': 16000,
    'earphone': 350
}

for price in product.values():
    price = price - (price * (20/100))
    print(price)
```

```python
product = {
    'laptop': 63000,
    'smartphone': 16000,
    'earphone': 350
}

for product, price in product.items():
    disPrice = price - (price * (20/100))
    print(f'{product} after 20% discount: {disPrice}')
```

```python
products = {
    'laptop': 63000,
    'smartphone': 16000,
    'earphone': 350
}

for product in enumerate(products):
    print(product)

    
```

```python
products = {
    'laptop': 63000,
    'smartphone': 16000,
    'earphone': 350
}

for index, product in enumerate(products):
    print(index, product)
```

```python
products = {
    'laptop': 63000,
    'smartphone': 16000,
    'earphone': 350
}

for price in enumerate(products.values()):
    print(price)

for index, price in enumerate(products.values()):
    print(index, price)
```

```python
products = {
    'laptop': 63000,
    'smartphone': 16000,
    'earphone': 350
}

for index, product in enumerate(products.items()):
    print(index, product)
```

```python
products = {
    'laptop': 63000,
    'smartphone': 16000,
    'earphone': 350
}

for index, product in enumerate(products.items(), 1):
    print(index, product)
```
