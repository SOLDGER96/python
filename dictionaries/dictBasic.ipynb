#### In Python, dictionaries are built-in data structures that store collections of key-value pairs. They work very similarly to real dictionaries, where you search for a word to find its corresponding meaning.

```python
dictionary = { 
    'key1': 'value1', 
    'key2': 'value2'
}
```

```python
pizza = {
    'name': 'Gawade Pizza Corner',
    'price': 220,
    'calPerSlice': 169,
    'topping': ['olive', 'mozzarella', 'tomato']
}

print(pizza['name'])
print(pizza['topping'])

print(pizza.get('name'))
print(pizza.get('topping'))

print('\n   # Keys of pizza')
print(pizza.keys())

print('\n   # Values of pizza')
print(pizza.values())

print('\n   # Items of pizza')
print(pizza.items())
```

#### The .clear() method removes all the key-value pairs from the dictionary:

```python
pizza = {
    'name': 'Gawade Pizza Corner',
    'price': 220,
    'calPerSlice': 169,
    'topping': ['olive', 'mozzarella', 'tomato']
}

print(pizza.clear())
```

```python
pizza = {
    'name': 'Gawade Pizza Corner',
    'price': 220,
    'calPerSlice': 169,
    'topping': ['olive', 'mozzarella', 'tomato']
}

pizza.pop('price')
print(pizza)
print(pizza.keys())
```

####  The .update() method updates the key-value pairs with the key-value pairs of another dictionary. If they have keys in common, their values are overwritten.

```python
pizza = {
    'name': 'Gawade Pizza Corner',
    'price': 220,
    'calPerSlice': 169,
    'topping': ['olive', 'mozzarella', 'tomato']
}

pizza.update({'price': 180, 'gst': '5%'})
print(pizza.keys())
print(pizza.values())
```
